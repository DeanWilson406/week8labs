```python
gh_turnover = pd.read_csv("gh_turnover_four_projects.csv") # Open the file
gh_turnover.info()
```

### Renaming Columns

```python
gh_turnover = gh_turnover.rename(columns = 
                                 {
                                     'domain':'project_domain',
                                     'language': 'project_language',
                                     'windows': 'project_age',
                                     'window_idx': 'quarter',
                                     'num_team': 'team_size',
                                     'num_commits': 'project_commits',
                                     'blau_gender': 'gender_blau',
                                     'Gini_gh_ten': 'gh_tenure_gini',
                                     'Core1': 'core_dev',
                                     'commits': 'user_commits',
                                     'propCommits': 'commit_proportion',
                                     'github_tenure': 'gh_tenure',
                                     'leavesNextQ': 'leaves_next_q'
                                 }
                                ) # Selected Columns Renamed
gh_turnover.info() # Verifying Renaming
```

### Data Wrangling

```python
gh_turnover['user_commits'].mean() # Calculating Mean of user_commits
```

```python
gh_turnover['user_commits'].median() # Calculating Median of user_commits
```

```python
turnover_new = gh_turnover.filter(regex='(^p)|(g)') # Filtering for columns starting with p or containing g
turnover_new.info() # Verify filter
```

```python
turnover_new.query('project_commits > 9000') # Selecting rows where project_commits is more than 9,000
```

```python
turnover_newer = ( 
    gh_turnover 
    .filter(regex='(^p)|(g)')
    .query('project_commits > 9000')
) # Chaining the filter and query in turnover_newer df
turnover_newer # Verify changes
```

```python
turnover_df1 = (
    gh_turnover
    .filter(regex='(e$)|(s)')
    .query('gh_tenure >= 1991')
) # Created new DF from gh_turnover that filters for columns ending in e and containing s, as well as selecting rows with gh_tenure greater than or equal to 1,991
turnover_df1 # Verify changes
```

```python
print(turnover_df1['user_commits'].mean()) # Print filtered user_commits mean
print(turnover_df1['user_commits'].median()) # Print filtered user_commits median
```

### Results
* The mean and median of turnover_df1 are both 3.5 and differ significantly from the earlier mean of 339.37 and median of 18. This is due to turnover_df1 being filtered to rows where gh_tenure is greater than or equal to the value of 1,991. This resulted in only two rows where user commits were 6 and 1. The sample size has drastically dropped, explaining the massive difference between the old and new mean and median.

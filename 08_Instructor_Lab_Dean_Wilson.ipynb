```python
 gh_teams.filter(['Team_type', 'human_work', 'work_per_human']) # Filtering Columns from gh_teams
```

```python
gh_teams.filter(regex = "t$") # Filtering Columns for last letter of "t"
```

```python
gh_teams.sort_values(by = ['Team_size_class', 'human_work', 'work_per_human'], ascending =False) #Sorting Data descending 
```

```python
gh_teams.query("Team_type == 'human-bot' and bot_members_count >= 3") # Selecting human-bot teams that have 3 or more bot_member_count
```

```python
gh_teams.query("Team_type == 'human' and Team_size_class == 'Large' and human_gini >= 0.75") # Finding large human teams that have .75 or more human_gini value
```

```python
gh_teams.query("Team_size_class == 'Small' or Team_size_class == 'Large'").shape # Finding 428 teams in small or large category
```

```python
gh_teams.query("(Team_size_class == 'Small' or Team_size_class == 'Large') and human_gini <= 0.25").shape # Finding small or large teams with .25 or less value in humann_gini 
```
